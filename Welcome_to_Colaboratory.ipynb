```
import numpy as np
import pandas as pd

# Set random seed for reproducibility
np.random.seed(42)

# Number of samples
num_samples = 1000

# Generate dummy features for claims
data = {
    'ClaimID': np.arange(1, num_samples+1),
    'ClaimAmount': np.random.uniform(1000, 20000, num_samples),  # Claim amounts
    'CustomerAge': np.random.randint(18, 75, num_samples),  # Customer age
    'CustomerTenure': np.random.randint(1, 30, num_samples),  # Years with the insurance company
    'IncidentHour': np.random.randint(0, 24, num_samples),  # Hour of incident
    'IsFraud': np.random.choice([0, 1], num_samples, p=[0.85, 0.15])  # Fraud labels (0 for genuine, 1 for fraud)
}

# Create DataFrame
df = pd.DataFrame(data)

# Display dataset
df.head()
```

```
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# Features and target variable
X = df[['ClaimAmount', 'CustomerAge', 'CustomerTenure', 'IncidentHour']]
y = df['IsFraud']

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Scale the features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```
import xgboost as xgb
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

# Create the XGBoost model
model = xgb.XGBClassifier(use_label_encoder=False, eval_metric='logloss')

# Train the model
model.fit(X_train_scaled, y_train)

# Predict on test data
y_pred = model.predict(X_test_scaled)

# Evaluate the model
print("Classification Report:\n", classification_report(y_test, y_pred))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
print("Accuracy Score:", accuracy_score(y_test, y_pred))
```

```
!pip install shap
import shap

# Create SHAP explainer
explainer = shap.Explainer(model, X_train_scaled)

# Explain the model's predictions for the test data
shap_values = explainer(X_test_scaled)

# Plot SHAP summary plot for feature importance
shap.summary_plot(shap_values, X_test, feature_names=X.columns)
```

```
# Simulate a real-time claim (new incoming data)
new_claim = np.array([[15000, 45, 5, 14]])  # Claim Amount, Customer Age, Customer Tenure, Incident Hour
new_claim_scaled = scaler.transform(new_claim)

# Predict fraud for this new claim
is_fraud = model.predict(new_claim_scaled)
probability = model.predict_proba(new_claim_scaled)

print(f"Is this claim fraudulent? {'Yes' if is_fraud[0] == 1 else 'No'}")
print(f"Fraud Probability: {probability[0][1]:.2f}")

# SHAP explanation for this prediction
shap_values_new_claim = explainer(new_claim_scaled)
shap.waterfall_plot(shap_values_new_claim[0])
```
